```python
#PCA is a dimention reduction method that is often use to reduce the dimenitonality of large data set by transformin g 
# a large set of variables into a smaller one that still contains most of the information in the large set.
#Int his tecqnique variables are transformed into a set of variables which are linear combination of original variabels . 
#This new set of variables are known as principal components .
#The first principal componet captures the most varaiational component in the data , but 2nd captueres the maximum variance that is orthogonal component and so on.



```

```python
#import necasary libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

df=pd.read_csv("iris.csv")
df
```

```python
df.species.replace({'Setosa':0,'Versicolor':1,'Virginica':2},inplace=True)
```

```python
df.head()
```

```python
# Standardize the data
features=['sepal_length','sepal_width','petal_length','petal_width']
```

```python

x=df.loc[:,features].values
y=df.loc[:,['species']].values
x=StandardScaler().fit_transform(x)
print(pd.DataFrame(data=x,columns=features).head())
```

```python
from sklearn.decomposition import PCA
#This line imports the pca class from the sckit-learn library which provides tools for dimention redction using pca
pca=PCA()\
#this line creates an instance of the pca class
#The PCA () contractor can take parameters like n_components to specify the number of pc to retain, but in this case 
X_new=pca.fit_transform(x)
```

```python
# Variance 
#explained variance for each principal component

explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
#PCA projecting to 2D
pca=PCA(n_components=2)
principalComponents=pca.fit_transform(x)
principalDF=pd.DataFrame(data=principalComponents,columns=['PC1','PC2'])
print(principalDF.head(150))
```
